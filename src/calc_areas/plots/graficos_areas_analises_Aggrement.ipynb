## funções de PLOTs

```python
def plot_pie_areasEstaveis(df_tmp, titlePlot, dict_color, nameLegend):
    fig = px.pie(
        df_tmp, 
        values='area', 
        names='classe',        
        hover_data=['name_class'],
        hole= 0.6,
        color='classe',
        color_discrete_map= dict_color
    )
    fig.update_traces(
        textposition='outside', # "inside"
        textinfo='label+percent', # value + 
        textfont_size= 16,
        marker=dict(line=dict(color='#000000', width=0))
    )
    fig.update_layout(
        # title= dict(
        #         text= titlePlot, 
        #         font=dict(size=20), 
        #         automargin=True, 
        #         yref='paper'),
        width = 500,
        height = 500,
        showlegend= False,
        # legend_title= nameLegend,
        # legend_title_font_size=15
    )
    fig.show()
    titlePlot = titlePlot.replace(" ", "_")
    print("save as ", titlePlot)
    fig.write_image("imagePPT/" + titlePlot + '.png')
```

## Fazer graficos de pizza para anos especificos no loop

```python
year = 2018
for cc, mapaAnalise in enumerate(lstnameImgCorre[:]):
    cc += 1
    lstDF = []
    classe = mapaAnalise.split("_")[-1]
    for npath in lstPath:
        if mapaAnalise in npath:
            # print("#", cc,  "  path >> ", npath)
            df_tmp = pd.read_csv(npath)
            df_tmp = df_tmp.drop(['system:index','.geo', 'bacia'], axis=1)
            #print("columnas dfTable" , df_tmp.columns)
            #print(df_tmp.head())
            lstDF.append(df_tmp)
            #os.remove(npath)
    if len(lstDF) > 0:
        concat_df  = pd.concat(lstDF, axis=0, ignore_index=True)
        print("temos {} filas ".format(concat_df.shape))
        dfGroup = concat_df.groupby(['classe', 'year'])['area'].sum().reset_index()
        print("agrupando por classe temos " , dfGroup.shape)
        dfGroup['classe'] = dfGroup['classe'].astype(str)
        # df['col1'].replace(di, inplace=True)
        dfGroup['name_class'] = dfGroup['classe'].replace(dict_name_aggrement, inplace=True)
        # dfGroup['classe'] = dfGroup['classe'].astype(int)
        print(" ",dfGroup.head())
        nomeLegend = "class Aggrements"
        namePlot = "Porcentagem de áreas de Aggrements, Classe " + classe + ", ano " + str(year) 
        plot_pie_areasEstaveis(dfGroup[dfGroup['year'] == year], namePlot, dict_color_aggrement, nomeLegend)
```

## Analises com a Serie historica 

```python
def plot_Serie_areasEstaveis(df_tmp, titlePlot, dict_color, nameLegend):
    figBar = px.bar(
        df_tmp, 
        x='year', 
        y ='area',
        color='classe', 
        color_discrete_map= dict_color,
        barmode='stack'
    )
    figBar.update_layout(
        # title= dict(
        #         text= titlePlot, 
        #         font=dict(size=20), 
        #         #automargin=True, 
        #         yref='paper',
        #         x= 0.02, xanchor="left",
        #         y= 1.0, yanchor="bottom", 
        # ),
        width = 1100,
        height = 400,
        legend_title=nameLegend,
        legend_title_font_size=15,
        yaxis=dict(
            title='Área em hectare (milhões)',
            titlefont_size=16,
            tickfont_size=14,
        ),
        xaxis=dict(
            title='Anos da serie historica',
            titlefont_size=16,
            tickfont_size=14,
        ),
    )
    figBar.show()
    titlePlot = titlePlot.replace(" ", "_")
    print("save as ", titlePlot)
    figBar.write_image("imagePPT/" + titlePlot + '.png')
```

```python
def plotArea_Serie_areasEstaveis(df_tmp, titlePlot, dict_color, nameLegend):
    figBar = px.area(
        df_tmp, 
        x='year', 
        y ='area',
        color='classe', 
        hover_data=['name_class'],
        color_discrete_map= dict_color,
        symbol= 'points',
        # barmode='stack'
    )
    figBar.update_layout(
        # title= dict(
        #         text= titlePlot, 
        #         font=dict(size=20), 
        #         #automargin=True, 
        #         yref='paper',
        #         x= 0.02, xanchor="left",
        #         y= 1.0, yanchor="bottom", 
        # ),
        width = 900,
        height = 400,
        legend_title=nameLegend,
        legend_title_font_size=15,
        yaxis=dict(
            title='Área em hectare (milhões)',
            titlefont_size=16,
            tickfont_size=14,
        ),
        xaxis=dict(
            title='Anos da serie historica',
            titlefont_size=16,
            tickfont_size=14,
        ),
        legend=dict(
            orientation="h",
            yanchor="bottom",
            y=-0.4,
            xanchor="right",
            x=0.85
        )
    )
    figBar.show()
    titlePlot = titlePlot.replace(" ", "_")
    print("save as ", titlePlot)
    # figBar.write_image("imagePPT/" + titlePlot + '.png')
```
